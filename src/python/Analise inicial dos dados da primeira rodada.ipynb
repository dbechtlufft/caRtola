```python
import pandas as pd
display.max_columns = 20
```

```python
df = pd.read_csv(PATH + 'rodada-2.csv')
#df.drop(columns = "Unnamed: 0", inplace = True)
```

```python
df.to_csv(PATH + 'rodada-2.csv', index = False)
```

```python
df.columns
```

```python
columns = ['Nome', 'Slug', 'Apelido', 'Foto', 'ID', 'Rodada', 'ClubeID', 'Posição', 'Status', 'Pontos', 'Preço', 
          'Variação', 'Media', 'Clube', 'FC', 'FS', 'PE', 'RB', 'SG', 'CA', 'FF','FD', 'G', 'I', 'A', 'GS', 'DD', 
           'FT', 'PP', 'CV', 'GC', 'DP']
df.columns = columns
df.columns
```

```python
cols_scouts_def = ['CA','CV','DD','DP','FC','GC','GS','RB','SG'] # alphabetical order
cols_scouts_atk = ['A','FD','FF','FS','FT','G','I','PE','PP'] # alphabetical order
cols_scouts = cols_scouts_def + cols_scouts_atk
```

```python
print("Dimensões originais dos dados: ", df.shape)

# remove todas as linhas cujo scouts são NANs 
df_clean = df.dropna(how='all', subset=cols_scouts)
print('qtde. de jogadores com scouts: ', df_clean.shape[0])

# remove técnicos e jogadores sem posição
df_clean = df_clean[(df_clean['Posição'] != "tec") & (~df_clean['Posição'].isnull())]
print("qtde. de linhas com posições válidas: ", df_clean.shape[0])

# atualiza nomes dos jogadores sem ids e remove jogadores sem nome
df_ids =  df.groupby('ID')['Apelido'].unique()
dict_ids = dict(zip(df_ids.index, [str(v[-1]) for v in df_ids.values]))
dict_ids = {k:v for k,v in dict_ids.items() if v != 'nan'}
df_clean['Apelido'] = df_clean['ID'].map(dict_ids)
df_clean = df_clean[~df_clean['Apelido'].isnull()]
print("qtde. de jogadores com nome: ", df_clean.shape[0])

# preenche NANs restantes com zeros (verificado antes!)
df_clean.fillna(value=0, inplace=True)

print("Dimensão dos dados após as limpezas: ", df_clean.shape)
df_clean.head(10)
```

```python
df_clean.head(5)
```

```python
df_util = df_clean[['Apelido','Posição', 'Status', 'Pontos', 'Preço', 'Variação', 'Media', 'Clube', 'CA', 'FF', 'FS', 'G', 'PE', 'RB', 'DD', 'GS', 'FC', 'FD', 'A', 'I', 'SG', 
           'FT', 'CV', 'GC', 'DP', 'PP']]
df_util.head(5)
```

```python
new_cols_scouts_atk = ['Apelido', 'Posição', 'Media', 'G','A','FD', 'FT', 'FF','FS','I','PE','PP'] # alphabetical order
pos = ['ata','mei']
df_ata_mei = df_clean[new_cols_scouts_atk][df_clean.Posição.isin(pos)]
df_ata_mei.head(20)
```

```python
df_ata_mei['F'] = df_ata_mei['FF'] + df_ata_mei['G'] + df_ata_mei['FT'] + df_ata_mei['FD']
df_ata_mei.head()
```

```python
df_ata_mei.sort_values(['F'], ascending = False).head(15)
```

```python
df_ata_mei['Efetividade'] = df_ata_mei['G'] / df_ata_mei['F']
df_ata_mei.sort_values(['Efetividade'], ascending = False).head(30)
```
